```python
import pandas as pd
from sqlalchemy import create_engine

# Database credentials
username = 'root'
password = '5555'
host = '127.0.0.1'
port = '3306'
database = 'traffic_logs'
csv_path = 'cleaned_traffic_stops' 

# Create SQLAlchemy engine
engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}")

chunksize = 5000     

# Loop through 
first_chunk = True
try:
    for i, chunk in enumerate(pd.read_csv('cleaned_traffic_stops', chunksize=chunksize)):
        if_exists_mode = 'replace' if first_chunk else 'append'
        chunk.to_sql(
            'traffic_stops',
            con=engine,
            if_exists=if_exists_mode,
            index=False
        )
        print(f"Inserted chunk {i+1} ({len(chunk)} rows)")
        first_chunk = False
    print(" All data inserted successfully into 'traffic_stops' table.")
except Exception as e:
    print(" Failed to insert data into MySQL:", e)
```
