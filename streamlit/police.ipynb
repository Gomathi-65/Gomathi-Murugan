```python
CREATE TABLE traffic_stops (
    stop_date DATE NOT NULL,                         -- Date of stop
    stop_time TIME,                                  -- Time of stop
    country_name VARCHAR(100),                       -- Place of stop
    driver_gender VARCHAR(10),                       -- Drivers gender
    driver_age INT,                                  -- Cleaned age
    driver_race VARCHAR(50),                         -- Drivers race/ethnicity
    violation VARCHAR(100),                          -- Type of violation
    search_conducted BOOLEAN DEFAULT FALSE,          -- Was a search conducted?
    search_type VARCHAR(50),                         -- Type (Frisk, Vehicle, etc.)
    stop_outcome VARCHAR(50),                        -- Outcome (Warning, Arrest, etc.)
    is_arrested BOOLEAN DEFAULT FALSE,               -- Was the driver arrested?
    stop_duration VARCHAR(20),                       -- Duration (e.g., '5 Min')
    drugs_related_stop BOOLEAN DEFAULT FALSE,        -- Drug-related stop?
    vehicle_number VARCHAR(20),                      -- Vehicle identifier
);
```
